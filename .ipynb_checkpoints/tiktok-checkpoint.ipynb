```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pymongo
import datetime
from flask import Flask
from sqlalchemy import create_engine
import json
from pathlib import Path  
```

```python
tiktok_csv = "Resources/tiktok.csv"
tiktok_df = pd.read_csv(tiktok_csv)
todrop = ['duration_mins',
'artist_id',
'key',
'speechiness',
'acousticness',
'instrumentalness',
'genre',
'playlist_id',
'playlist_name',
'album_id']
tiktok_df = tiktok_df.drop(columns=todrop)
tiktok_df = tiktok_df.drop_duplicates('track_id')
tiktok_df
```

```python
apple_music = "Resources/audd_music_apple_music.csv"
apple_music_df = pd.read_csv(apple_music)
todrop = ['url',
'discNumber',
'durationInMillis',
'trackNumber',
'composerName',
'artwork.width',
'artwork.height',
'artwork.url',
'artwork.bgColor',
'artwork.textColor1',
'artwork.textColor3',
'artwork.textColor4',
'artwork.textColor2',
'albumName',
'playParams.id',
'playParams.kind']
apple_music_df = apple_music_df.drop(columns = todrop)
apple_music_df = apple_music_df.rename(columns={"isrc": "apple_music.isrc"})
apple_music_df.head()
```

```python
spotify_music = "Resources/audd_music_spotify_music.csv"
spotify_music_df = pd.read_csv(spotify_music)
todrop = ['is_playable',
'linked_from', 
'available_markets', 
'disc_number', 
'duration_ms', 
'href', 
'preview_url', 
'track_number', 
'uri', 
'album.name', 
'name', 
'album.album_group', 
'album.album_type', 
'album.id', 
'album.uri', 
'album.available_markets', 
'album.href', 
'album.images', 
'album.external_urls.spotify', 
'album.release_date', 
'album.release_date_precision', 
'external_ids.isrc', 
'external_urls.spotify', 
'artists_ids', ]
spotify_music_df = spotify_music_df.drop(columns = todrop)
spotify_music_df = spotify_music_df.rename(columns={"id": "spotify.id"})
spotify_music_df = spotify_music_df.rename(columns={"popularity": "spotify_popularity"})
spotify_music_df.head()
```

```python
music = "Resources/audd_music.csv"
music_df = pd.read_csv(music)
todrop = ['id',
'song_link',
'label']
music_df = music_df.drop(columns = todrop)
music_df = music_df.rename(columns={"title": "track_name"})
music_df.head()
```

```python
billboard_charts = "Resources/charts.csv"
billboard_charts_df = pd.read_csv(billboard_charts)
todrop = ['date',
'rank',
'artist',
'last-week',
'weeks-on-board']
billboard_charts_df = billboard_charts_df.drop(columns = todrop)
billboard_charts_df = billboard_charts_df.drop_duplicates()
billboard_charts_df = billboard_charts_df.rename(columns={"song": "track_name"})
billboard_charts_df
```

```python
music_tiktok_df = pd.merge(tiktok_df, music_df, how = 'inner', on = 'track_name')
music_tiktok_df = music_tiktok_df.drop_duplicates()
music_tiktok_df = music_tiktok_df.dropna()
music_tiktok_df
```

```python
music_tiktok__spotify_df = pd.merge(music_tiktok_df, spotify_music_df, how = 'inner', on = 'spotify.id')
music_tiktok__spotify_df
```

```python
music_tiktok__spotify_apple_df = pd.merge(music_tiktok__spotify_df, apple_music_df, how = 'inner', on = 'apple_music.isrc')
music_tiktok__spotify_apple_df = music_tiktok__spotify_apple_df.drop_duplicates('track_name') 
music_tiktok__spotify_apple_df = music_tiktok__spotify_apple_df.dropna()
music_tiktok__spotify_apple_df
```

```python
billboard_tiktok_df = pd.merge(music_tiktok__spotify_apple_df, billboard_charts_df, how = 'inner', on = 'track_name')
billboard_tiktok_df = billboard_tiktok_df.drop_duplicates('track_name') 
billboard_tiktok_df = billboard_tiktok_df.dropna()
billboard_tiktok_df.head()
```
