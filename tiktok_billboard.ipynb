```python
#Importing all the Library that will be needed.

import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import sweetviz as sv
```

## Extract and Transform Data

```python
#converting csv file into dataframe using pandas.

tiktok_csv = "Resources/tiktok.csv"
tiktok_df = pd.read_csv(tiktok_csv)

#converting name into all lower case to ensure that track_name can be joined. 
tiktok_df['track_name'] = tiktok_df['track_name'].str.lower().str.strip()

#droppring attribute not used for analysis
todrop = ['duration',
'artist_id',
'key',
'genre',
'playlist_id',
'playlist_name',
'album_id',
'release_date']

#Drop any duplicated 'track_name'
tiktok_df = tiktok_df.drop(columns=todrop)
tiktok_df = tiktok_df.drop_duplicates('track_name')

#Check to make sure that there is no duplicate on the track_name
tiktok_df['track_name'].value_counts()
```

```python
music = "Resources/audd_music.csv"
music_df = pd.read_csv(music)

music_df['track_name'] = music_df['title'].str.lower().str.strip()
todrop = ['id',
'song_link',
'label',
'release_date',
'title',
'artist']
music_df = music_df.drop(columns = todrop)
music_df = music_df.drop_duplicates('track_name')
```

```python
apple_music = "Resources/audd_music_apple_music.csv"
apple_music_df = pd.read_csv(apple_music)
todrop = ['url',
'discNumber',
'durationInMillis',
'trackNumber',
'name',
'composerName',
'artwork.width',
'artwork.height',
'artwork.url',
'artwork.bgColor',
'artwork.textColor1',
'artwork.textColor3',
'artwork.textColor4',
'artwork.textColor2',
'albumName',
'artistName',
'playParams.id',
'playParams.kind']

apple_music_df = apple_music_df.drop(columns = todrop)

#renaming the column to allow dataframes to be merged.
apple_music_df = apple_music_df.rename(columns={"isrc": "apple_music.isrc"})
apple_music_df = apple_music_df.drop_duplicates('apple_music.isrc')
```

```python
apple_music_df['apple_music.isrc'].value_counts()
```

```python
spotify_music = "Resources/audd_music_spotify_music.csv"
spotify_music_df = pd.read_csv(spotify_music)
todrop = ['is_playable',
'linked_from', 
'available_markets', 
'disc_number', 
'duration_ms', 
'href', 
'preview_url', 
'track_number', 
'uri', 
'album.name', 
'name', 
'album.album_group', 
'album.album_type', 
'album.id', 
'album.uri', 
'album.available_markets', 
'album.href', 
'album.images', 
'album.external_urls.spotify', 
'album.release_date', 
'album.release_date_precision', 
'external_ids.isrc', 
'external_urls.spotify', 
'artists_ids', ]
spotify_music_df = spotify_music_df.drop(columns = todrop)
spotify_music_df = spotify_music_df.rename(columns={"id": "spotify.id"})
spotify_music_df = spotify_music_df.rename(columns={"popularity": "spotify_popularity"})
spotify_music_df = spotify_music_df.drop_duplicates('spotify.id')
spotify_music_df
```

```python
#Check to make sure that there is no duplicate on the 'spotify.id'
spotify_music_df['spotify.id'].value_counts()
```

```python
billboard_charts = "Resources/charts.csv"
billboard_charts_df = pd.read_csv(billboard_charts)
todrop = ['date',
'rank',
'artist',
'last-week',
'weeks-on-board']
billboard_charts_df = billboard_charts_df.drop(columns = todrop)
billboard_charts_df = billboard_charts_df.rename(columns={"song": "track_name"})
billboard_charts_df['track_name'] = billboard_charts_df['track_name'].astype('string')
billboard_charts_df['track_name'] = billboard_charts_df['track_name'].str.lower().str.strip()
billboard_charts_df = billboard_charts_df.drop_duplicates('track_name')
billboard_charts_df
```

```python
#Merged all the different dataframe to create dimensional table

tiktok_music_df = tiktok_df.merge(music_df, how = 'inner', on = 'track_name')
tiktok_music_apple_df = tiktok_music_df.merge(apple_music_df, how = 'inner', on = 'apple_music.isrc')
tiktok_music_apple_spotify_df = tiktok_music_apple_df.merge(spotify_music_df, how = 'inner', on ='spotify.id')
tiktok_dimension_df = tiktok_music_apple_spotify_df.merge(billboard_charts_df, how = 'left', on = 'track_name')
```

```python
tiktok_dimension_df
```

```python
#Show the datatype of each columns 
tiktok_dimension_df.dtypes
```

### Convert Datatypes

```python
tiktok_dimension_df["releaseDate"] = pd.to_datetime(tiktok_dimension_df["releaseDate"])
```

```python
#Using Sweetviz to generate statistical visualization for each column.
my_report = sv.analyze(tiktok_dimension_df)
my_report.show_html("tiktok_report.html") # Default arguments will generate to "SWEETVIZ_REPORT.html"
```

```python
tiktok_dimension_df = tiktok_dimension_df.rename(columns = {"apple_music.isrc" : "apple_music_id", "spotify.id" : "spotify_id", "peak-rank": "peak_rank", "genreNames": "genrenames", "releaseDate": "releasedate"})
tiktok_dimension_df
```

### Create database conncetion

```python
connection_string = "postgres:postgres@localhost:5432/tiktok_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

### Load DataFrames into DB

```python
df = tiktok_dimension_df [['track_id', 'track_name', 'artist_name', 'popularity', 'danceability',
       'energy', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_mins',
       'album', 'apple_music_id', 'spotify_id', 
       'releasedate', 'spotify_popularity', 'explicit', 'peak_rank']]
```

```python
df.to_sql(name='tiktok_billboard', con=engine, if_exists='append', index = False)
```
